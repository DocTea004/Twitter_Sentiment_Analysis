```python
#Remove Handles since they do not contribute to the sentiment analysis
def remove_pattern(text,pattern):
    
    # re.findall() finds the pattern i.e @user and puts it in a list for further task
    r = re.findall(pattern,text)
    
    # re.sub() removes @user from the sentences in the dataset
    for i in r:
        text = re.sub(i,"",text)
    
    return text
```

```python
Combined_DataSet['Dexoangle'] = np.vectorize(remove_pattern)(Combined_DataSet['tweet'], "@[\w]*")
Combined_DataSet.head()
```

```python
#Remove punctuations and other symbols which do not help in the analysis
```

```python
Combined_DataSet['Dexoangle'] = Combined_DataSet['Dexoangle'].str.replace("[^a-zA-Z#]", " ")

Combined_DataSet.head(10)
```

```python
#Select the length of words to use
```

```python
Combined_DataSet['Dexoangle'] = Combined_DataSet['Dexoangle'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

Combined_DataSet.head(10)
```

```python
#Convert Strings To Tokens
```

```python
Token_Tweets = Combined_DataSet['Dexoangle'].apply(lambda x: x.split())

Token_Tweets.head()
```

```python
#We need to strip of the prefixes(Stemming)
#Strip of prefixes such as s,es,ing,ly,etc
```

```python
from nltk import PorterStemmer
Port_Stm = PorterStemmer()
Token_Tweets = Token_Tweets.apply(lambda x: [Port_Stm.stem(y) for y in x])
Token_Tweets.head()
```

```python
#Stitch tokens Back
```

```python
for y in range(len(Token_Tweets)):
    Token_Tweets[y] = ' '.join(Token_Tweets[y])

Combined_DataSet['Dexoangle'] = Token_Tweets
Combined_DataSet.head()
```

```python
#Extract HashTags
```

```python
def Hashtag_Extractor(m):
    Hashtags=[]
    
    
    for i in m:
        Hash_Tags = re.findall(r'#(\w+)',i)
        Hashtags.append(Hash_Tags)
    
    return Hashtags
```

```python
#Positive HashTags
```

```python
Positive_HashTags = Hashtag_Extractor(Combined_DataSet['Dexoangle'][Combined_DataSet['label']==0])
Positive_HashTags
```

```python
#Unset all positive HashTags
```

```python
Unset_PositiveTags = sum(Positive_HashTags,[])
Unset_PositiveTags
```

```python
#Negative HashTags
```

```python
Negative_HashTags = Hashtag_Extractor(Combined_DataSet['Dexoangle'][Combined_DataSet['label']==1])
Negative_HashTags
```

```python
#Unset Negative HashTags
```

```python
Unset_NegativeHashTag = sum(Negative_HashTags,[])
Unset_NegativeHashTag
```

```python
#Counting Frequency of Words Having Positive Sentiments
```

```python
Positive_Frequency = nltk.FreqDist(Unset_PositiveTags)
Positive_Frequency
```

```python
#Data Frame for Positive Sentiments
```

```python
Positive_Dataframe = pd.DataFrame({'Hashtags':list(Positive_Frequency.keys()),'Count':list(Positive_Frequency.values())})
Positive_Dataframe.head(10)
```

```python
#Using Bag of Words which is a method used to extract features from text documents. 
#It creates a vocabulary of all the unique words occurring in all the documents in the training set.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words='english')
bow = bow_vectorizer.fit_transform(Combined_DataSet['Dexoangle'])

df_bow = pd.DataFrame(bow.todense())

df_bow
```

```python
#TD-IF
```

Using TF-IDF TF-IDF stands for term frequency-inverse document frequency, TF-IDF weight is a weight often used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is to a document in a collection. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(max_df=0.90, min_df=2,max_features=1000,stop_words='english')

tfidf_matrix=tfidf.fit_transform(Combined_DataSet['Dexoangle'])

tfidf_Dataframe = pd.DataFrame(tfidf_matrix.todense())

tfidf_Dataframe
```
